```python
import json
import toml
import os
from dotenv import load_dotenv

load_dotenv()
creds_path = os.getenv("GCP_CREDENTIALS_PATH")
```

```python
# Open the file in read mode ('r')
with open(creds_path, 'r') as file:
    # Deserialize the JSON data from the file into a Python object
    json_data = json.load(file)
     
with open('../streamlit/.streamlit/secrets.toml', 'w') as file:
    toml.dump({"gcp_service_account": json_data}, file)
```
